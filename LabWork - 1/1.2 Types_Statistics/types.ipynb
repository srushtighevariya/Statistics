# Types of Statistics

# Q-1

1. Descriptive Statistics and Inferential Statistics
    - Descriptive Statistics : 
        - To summarize and describe the main features of a dataset (mean,median,mode,).
        - They are a fundamental first step in data analysis, focusing on describing what the data shows rather than making predictions about a larger population (like inferential statistics do). 
    - Inferential Statistics :
        - To make predictions or inferences about a population based on a sample(e.g., hypothesis testing, confidence intervals).
        - Inferential statistics is a branch of statistics that uses sample data to make predictions, generalizations, and draw conclusions about a larger population, moving beyond just describing the data itself.

2. Visual map or list to differenciate both
    - Statistics
        - Descriptive
            - Mean, Median, Mode
            - Range, Variance, Std Dev
            - Charts: Histogram, Boxplot
        - Inferential
             - Hypothesis Testing
             - Confidence Intervals
             - Regression, ANOVA

# Q-2

1. Load a sample dataset
2. use .describe() to get an overview
3. List the values return and explain each values

```python
import pandas as pd

# Load a sample dataset
data = {
    'Age': [22, 25, 30, 28, 35, 40, 29],
    'Score': [88, 92, 95, 85, 90, 87, 93]
}
df = pd.DataFrame(data)
print(df)

# using .describe()
print("\nDescription of the DataFrame:")
print(df.describe())


# Explaination:
print("\nExplanation of the output:")
print('count - number of non-null entries in each column')
print('mean - average value of each numeric column')
print('std - standard deviation of each numeric column')
print('min - minimum value of each numeric column')
print('25% - 25th percentile (first quartile) of each numeric column')
print('50% - 50th percentile (median) of each numeric column')
print('75% - 75th percentile (third quartile) of each numeric column')
print('max - maximum value of each numeric column')
```

# Q-3

1. Calculate the mean of the number columnn.
2. Calculate the median
3. Calculate the mode.
4. Compair the bar chart of mean, median and mode

```python
import numpy as np
import matplotlib.pyplot as plt

# Mean
mean1 = df['Score'].mean()
mean2 = np.mean(df['Score'])

# Median
median1 = df['Score'].median()
median2 = np.median(df['Score'])

# Mode
mode = df['Score'].mode()[0]

# Print values
print("Mean:", mean1, mean2)
print("Median:", median1, median2)
print("Mode:", mode)

# Bar chart
plt.bar(['Mean', 'Median', 'Mode'], [mean1, median1, mode], color=['red', 'black', 'blue'])
plt.title('Central Tendency Comparison')
plt.ylabel('Score')
plt.show()
```

# Q-4

1. Calculate the range 
2. Calculate variance 
3. Calculate Standard Deviation
4. Create a boxplot

```python
# Range
range_val = df['Score'].max() - df['Score'].min()

# Variance
var1 = df['Score'].var()
var2 = np.var(df['Score'], ddof=1)  # Sample variance

# Standard Deviation
std1 = df['Score'].std()
std2 = np.std(df['Score'])

print("Range: Spread between max and min : ", range_val)
print("Variance:  Average squared deviation ", var1, var2)
print("Standard Deviation: Average deviation from mean", std1, std2)

# Boxplot
plt.boxplot(df['Score'])
plt.title('Score Dispersion (Boxplot)')
plt.ylabel('Score')
plt.show()
```
