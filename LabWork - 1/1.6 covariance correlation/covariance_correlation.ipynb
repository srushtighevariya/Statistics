# Covariance

## Q-1

1. Genreate two list of sudetns scores for MAths and Physics
2. Convert to pandas dataframe
3. Calculate the covariance using .cov()
4. Calculate Pearson correlation coefficient using.corr(methos='pearson')
5. Shuffle one of the score list randomly
6. Calculate Spearman rank correlation using .corr(methos='spearman')
7. Discuss the difference between Spearman and Pearson
    - Pearson measures linear correlation assuming normality.
    - Spearman measures rank correlation, robust to outliers and non-linear relationships.

```python
import pandas as pd
import numpy as np

math_scores = [88, 92, 75, 85, 90, 79, 95, 80, 87, 91]
physics_scores = [84, 89, 78, 82, 88, 76, 94, 77, 85, 90]


# 2. Convert to pandas dataframe
df = pd.DataFrame({'Math': math_scores, 'Physics': physics_scores})

# 3. Calculate the covariance using .cov()
covariance = df.cov()
print(covariance,end="\n\n")

# 4. Calculate Pearson correlation coefficient using.corr(methos='pearson')
pearson_corr = df.corr(method='pearson')
print(pearson_corr,end="\n\n")

# 5. Shuffle one of the score list randomly
np.random.shuffle(df['Physics'].values)

# 6. Calculate Spearman rank correlation using .corr(methos='spearman')
spearman_corr = df.corr(method='spearman')
print(spearman_corr,end="\n\n")
```

## Q-2

1. Generate and plot a Uniform Distribution using Numpy
2. Use the Histogram with 50 bins

```python
import matplotlib.pyplot as plt

uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=50, color='pink')
plt.title("Uniform Distribution")
plt.show()
```

3. Generate a Binomial Distribution using Numpy or scipy.stats.binom
4. Compare it visually with a Normal Distribution

```python
from scipy.stats import binom, norm

binom_data = binom.rvs(n=10, p=0.5, size=1000)
normal_data = np.random.normal(loc=5, scale=1.5, size=1000)

plt.hist(binom_data, bins=30, alpha=0.6, label='Binomial')
plt.hist(normal_data, bins=30, alpha=0.6, label='Normal')
plt.legend()
plt.title("Binomial vs Normal Distribution")
plt.show()
```

5. Simulate a Poission Distribution of daily customer counters
6. Plot and Interprete

```python
poisson_data = np.random.poisson(lam=20, size=1000)
plt.hist(poisson_data, bins=30, color='lightgreen')
plt.title("Poisson Distribution of Daily Customers")
plt.show()
```

## Q-3

1. Generate a population withe skewed distribution
2. Draw 1000 sample of size 30 and calculate sample means
3. plot the Histogram of sample mean


```python
population = np.random.exponential(scale=2.0, size=100000)

means_30 = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]
plt.hist(means_30, bins=50, color='green')
plt.title("Sample Means (n=30)")
plt.show()
```

4. Increase the samle size to 100 per sample
5. Show how CLT strengthens with larger sample size
    - As sample size increases, the distribution of sample means becomes more normal.
    - This demonstrates the Central Limit Theorem, even when the population is skewed.

```python
means_100 = [np.mean(np.random.choice(population, 100)) for _ in range(1000)]
plt.hist(means_100, bins=50, color='purple')
plt.title("Sample Means (n=100)")
plt.show()
```
