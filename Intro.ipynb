## Q-2

1. Structured vs Unstructured Data
    - Structured Data: 
        - Data organized in predefined formats like tables.
        - Organized in rows and columns (e.g., spreadsheets, SQL databases).
        - Storage -> Relational Database
        - easily searchable and predictable.
    - Unstructured Data: 
        - Data that does not have a predefined model or schema.
        - No predefined format (e.g., text, images, audio)
        - Storage -> NoSQL databases or Data Lakes.
        - requires processing (like NLP) to extract meaning.
2. Applications of Statastics.
    - Healthcare: Predicting disease outbreaks, analyzing treatment effectiveness and analyzing patient survival rates.
    - Business: Demand forecasting, risk analysis, Inventory management and supply chain optimization.
    - Marketing: Customer behavior analysis, campaign performance tracking, Segmenting customers into personas based on buying behavior.

## Q-3

1. Create a Pandas DataFrame with [Name(categorical), Age(numeric), Department(categorical), Salary(numeric)]
2. Use .dtypes to show types of each column.
3. Use select_dtype() to filter numeric & categorical data.

```python
import pandas as pd

#1. creating DataFrame
data ={'Name':['Srushti','Khushali','Aaditya','Krish','Tirth'],
       'Age':[27,24,35,30,35],
       'Department':['HR','IT','Finance','IT','Marketing'],
       'Salary':[50000,30000,75000,65000,90000]}

df = pd.DataFrame(data)


# 2. Show DataTyes of each column
print(f'Datatypes of each column : \n{df.dtypes}')

#3. FilterNumeric and Categorical data
numeric_data = df.select_dtypes(include=['number'])
categorical_data = df.select_dtypes(include=['object'])

print(f'\nNumeric Data : \n{numeric_data}')
print(f'\nCategorical Data : \n{categorical_data}')
```

1. Histogram for age distribution.

4. Boxplot for salary.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Histogram for age distribution.

sns.histplot(df['Age'], bins=6, kde=True)
plt.title('Age Distribution')

print('We can see younger one\'s age is 24 and most employees\'s age are between 32 to 36 .')
```

```python
# 2. Bar chart for department count.

sns.countplot(df['Department'])
plt.title('Department Count')

print('We can say that IT department has highest number of employees.')
```

```python
# 3. Pie chart of department percentage.

dept_counts = df['Department'].value_counts()
dept_counts.plot.pie(autopct='%2.2f%%', startangle=90)
plt.title('Department Percentage')

print('We can say that IT department holds the 40% employees.')
```

```python
# 4. Boxplot for salary.
sns.boxplot(df['Salary'])
plt.title('Salary Boxplot')

print('We can see that there is no outlier in salary and middle 50% Salary is in range of 50000 to 75000.')
```

## Q-5

1. Generate dataset of 1000 random numbers(test score) using Numpy
2. Take random sample of 100 entries.
3. Calculate and compare mean and standard deviation of : Entire population, Sample
4. Explain the difference between population and Sample statistics.
    - Population Statistics (Parameters): These are values calculated from the entire group you are interested in. They are usually fixed but often impossible to measure directly.
    - Sample Statistics: These are values calculated from a subset of the population. They are used as "estimates" for the population parameters. The difference between the sample statistic and the actual population parameter is known as sampling error.

```python
import numpy as np
```

```python
# 1. Generate dataset of 1000 random numbers(test score) using Numpy
np.random.seed(30)
population = np.random.normal(size=1000,loc=70, scale=10)

```

```python
# 2. Take random sample of 100 entries.
sample = np.random.choice(population,size=100)
```

```python
# 3. Calculate and compare mean and standard deviation of : Entire population, Sample
population_mean = np.mean(population)
population_sd = np.std(population)
sample_mean = np.mean(sample)
sample_sd = np.std(sample)

print(f'Mean : \n\tPopulation - {population_mean:.2f} \n\tSample - {sample_mean:.2f} \nStandard DEviation : \n\tPopulation - {population_sd:.2f} \n\tSample - {sample_sd:.2f}')
```
